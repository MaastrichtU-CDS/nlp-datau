```python
from mtranslate import translate
import traceback

def mtranslate_text(src_txt):
    retry = 0
    max_retry = 3
    result = None
    while not result and retry < max_retry:
        try:
            result = translate(src_txt, 'nl', 'en')
        except Exception as e:
            print("An exception was thrown!", src_txt)
            print(traceback.format_exc())
        retry += 1
    return result
```

```python
import os
from tqdm.notebook import tqdm

DIR_IN = '/home/jovyan/work/resources/data_ignored/processed/'
DIR_OUT = '/home/jovyan/work/resources/data_ignored/processed-nl/'

FILE_IN_SRC = DIR_IN + 'src-train.txt'
FILE_IN_TGT = DIR_IN + 'tgt-train.txt'

FILE_OUT_SRC = DIR_OUT + 'src-train.txt'
FILE_OUT_TGT = DIR_OUT + 'tgt-train.txt'

tgt_size = 10000000

try:
    os.mkdir(DIR_TGT)
except OSError as error:
    print(error)

def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1


writer_src = open(FILE_OUT_SRC, "w")
writer_tgt = open(FILE_OUT_TGT, "w")

line_count = 0
success_count = 0

num_lines = sum(1 for line in open(FILE_IN_SRC,'r'))        
with open(FILE_IN_SRC, 'r') as reader_src:
    with open(FILE_IN_TGT, 'r') as reader_tgt:
        
        for line in tqdm(reader_src, total=num_lines):
            line_tgt = reader_tgt.readline()
            if not line_tgt:
                break
            translation_src = mtranslate_text(line)
            translation_tgt = mtranslate_text(line_tgt)
            if translation_src and translation_tgt:
                success_count += 1
                writer_src.write(translation_src + "\n")
                writer_tgt.write(translation_tgt + "\n")
                if success_count%100 == 0:
                    print("line {}: {} {}".format(success_count, translation_src, translation_tgt))

reader_src.close()
reader_tgt.close()
writer_src.close()
writer_tgt.close()

print('items FILE_IN_SRC', str(file_len(FILE_IN_SRC)))
print('items FILE_IN_TGT', str(file_len(FILE_IN_TGT)))
print('items FILE_OUT_SRC', str(file_len(FILE_OUT_SRC)))
print('items FILE_OUT_TGT', str(file_len(FILE_OUT_TGT)))

```
